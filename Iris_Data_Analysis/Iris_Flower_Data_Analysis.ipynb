```python
data = pd.read_csv('Iris.csv')
```

## 2. Find statistics

```python
data.describe()
```

#### a. Find the mean of the SepalLengthCm.

```python
mesepal = data['SepalLengthCm'].mean().round(2)
mesepal
```

## 3. Grouping

#### a. Find the mean of all species.

```python
data.groupby('Species').mean().round(2)
```

## Plot the data.

```python
sns.scatterplot(x='SepalLengthCm', y = 'SepalWidthCm', hue = 'Species', edgecolor = 'red', data = data)
plt.title('Flower Size Distrubution')
plt.xlabel('Sepal Length (cm)')
plt.ylabel("Sepal Width (cm)")
plt.show()
```

```python
sns.scatterplot(x='PetalLengthCm', y = 'PetalWidthCm', hue = 'Species', edgecolor = 'red', data = data)
plt.title('Flower Size Distrubution')
plt.xlabel('Petal Length (cm)')
plt.ylabel("Petal Width (cm)")
plt.show()
```
