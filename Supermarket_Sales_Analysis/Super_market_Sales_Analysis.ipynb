## 1. Count values of each type of payment

```python
countpayment = data.value_counts('Payment')
```

```python
plt.bar(countpayment.index, countpayment.values)
plt.xlabel('Payment Method')
plt.ylabel('Count')
plt.title('Count of Each Payment Method')

plt.show()
```

## 2. Find the total sales of each city.

```python
sal = data.groupby('City')['Total'].sum().round(2)
sal
```

```python
plt.bar(sal.index, sal.values)
plt.xlabel("City Names")
plt.ylabel("Total Sales")
plt.title("Total Sales by City")
plt.show()
```

### 3.Find gender wise total shoping.

```python
gshp = data.groupby('Gender')['Total'].sum().round(2)
gshp
```

```python
plt.bar(gshp.index, gshp.values)
plt.xlabel('Gender')
plt.ylabel('Total shoping')
plt.title("Total shoping by gender wise")
plt.show()
```
