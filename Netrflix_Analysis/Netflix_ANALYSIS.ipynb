## 6. Find the percentage of movie and tv shows in whole data 

```python
tot = data['type'].value_counts().sum() # finding the total rows
tot
```

```python
moco = data['type'].str.contains('Movie').sum()  # finding the total count of movie rows, it can be done in various way like by len().
moco
```

```python
tvco = data['type'].str.contains('TV Show').sum()  # finding the total count of TV Show rows,
tvco
```

```python
movie_percentage = (moco/tot)*100   # finding the percentage of movie
movie_percentage.round()
```

```python
shows_percentage = (tvco/tot)*100   # finding the percentage of TV Show
shows_percentage.round()
```

```python
total_percentage = movie_percentage.round().astype(int),shows_percentage.round().astype(int)
total_percentage
```
